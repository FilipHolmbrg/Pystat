### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The Object Oriented Style (OOS) and pyplot "MATLAB" style. OOs is preferred.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

figure = the whole image "canvas" object that everything happens on. Keeps track of axes. "eller hela väggen"

artist = everything visible on a figure (t ex Figure, axes, axis objects) "allt vi slänger på figuren"

axes = A special type of artist that is attached to a Figure and includes usually 2-3 (x,y,z) axis objects. Provides scales for the data in the axes. Title, x-label and y-label are part of Axes. It is the primary entry point for OOs ploting. "tapeten"

axis = object set scale/limit/generate ticks/ticklabels on the Axis. "x/y axlar"

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OOs style
x = np.linspace(-4,4,100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('x-values')
ax.set_ylabel('y-values')
ax.set_title('OO-Style')

plt.figure()
plt.plot(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Pyplot "MATLAB" style')

plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
# x = np.linspace(-2, 2, 100)
# y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (5,6))
ax1.scatter(random_data_x,random_data_y)
ax2.bar(names, values)

plt.show()
```
